```python
class Appetizers:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price
    def __str__(self):
        return f"{self.name} \n{self.description} \n{self.price}"
Appetizers1 = Appetizers("Wings", "Sauces : Buffalo, Texas Sauce, Lemon", "$10")
Appetizers2 = Appetizers("Mozzarella Sticks", "Fried, Marinara (inculded) Sauce", "$8")

print(Appetizers1)
print(Appetizers2)
```
